View only the name,date and status of coins with highest or lowest & open or close values from dataset?

```python
status = cryptocurrency[["name","date","open","high","low","close"]]
print(status)
```

How to sort the values in open,high,low,close columns?

```python
status.sort_values(["open"],ascending = True)
```

Find the mean value of the columns 'high','open','low','close'?

```python
round(status[['high','open','low','close']].mean())
```

Find the max value of the columns 'high','open','low','close'?

```python
status[['high','open','low','close']].max()
```

Find the minimum value of the columns 'high','open','low','close'?

```python
status[['high','open','low','close']].min()
```

How to remove the repeated names of the coins from 'name'? 

```python
status_1 = status.drop_duplicates(subset = "name")

print(status_1.head())
```

```python
status_1.shape
```

```python
status_tail = status.drop_duplicates(subset = "name")
print(status_tail.tail())
```

How to find the stock values of 'name' with 'date' columns together?

```python
status_1.groupby(["name","date"]) [["high","low"]].max()
```

```python
status_1.groupby(["name","date"]) [["high","low"]].min().head()
```

How to change the index and set the values of dataframe in tabular representation?

```python
status_1.pivot_table(values = "high", index = "date", columns = "name")#replace()
```

```python
status_1.pivot_table(values = "low", index = "date", columns = "name")
```
